```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('clusters.csv')
df
```

Lets use PCA to create a 2-d representation


```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
scaled = minmax.fit_transform(df) #firstly scale data points
pca = PCA(n_components = 2)
df2_D = pca.fit_transform(scaled) # 2d representation
```

```python
plt.figure(figsize = (8,8))
x = sns.scatterplot(df2_D[:,0],df2_D[:,1])
x.set(title = 'Representation in two dimentions',xlabel = '1st dimention',ylabel = '2nd dimention')
plt.show()
```

At the 1st point of view it can be minumun something between 2-6 clusters. To find out the best option wcss have to be calculated

WCSS is sum of squared distances of samples to their closest cluster center, returned by inertia_ argument

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
#Plot the WCSS values onto a line graph
plt.plot(range(1, 11), wcss)
plt.title('WCSS by Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

There is a huge deacrease in WCSS from 1 to 2nd cluster from 2 to 3 and from 3 to 4. There is slightly no point of making more than 4 clusters

```python
from sklearn.cluster import KMeans

# Create a model based on 4 centroids
model = KMeans(n_clusters=4, init='k-means++', n_init=100, max_iter=1000)
# Fit to the data and predict the cluster assignments for each data point
km_clusters = model.fit_predict(df)
# View the cluster assignments
km_clusters
```

```python
#Lets plot it
def plot_clusters(samples, clusters):
    col_dic = {0:'blue',1:'green',2:'orange',3:'yellow'}
    mrk_dic = {0:'*',1:'x',2:'+',3:'o'}
    colors = [col_dic[x] for x in clusters]
    markers = [mrk_dic[x] for x in clusters]
    for sample in range(len(clusters)):
        plt.scatter(samples[sample][0], samples[sample][1], color = colors[sample], marker=markers[sample], s=100)
    plt.xlabel('Dimension 1')
    plt.ylabel('Dimension 2')
    plt.title('Assignments')
    plt.show()

plot_clusters(df2_D, km_clusters)
```

```python
labels = np.unique(km_clusters)
labels #final labels for our model
```
